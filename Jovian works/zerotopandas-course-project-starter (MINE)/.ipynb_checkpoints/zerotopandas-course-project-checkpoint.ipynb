## Downloading the Dataset

I have already download a dataset from `kaggle` in WinRar zip format, now I just need to `extract` it and `read` it.

Let's begin by extracting the data, and listing the files within the dataset, <br>
* We will be using library called `patool` to extract `winRaR zip` files

````python
# pip install patool
````

#### A function to extract and read the data

````python
from pathlib import Path  # to access paths
import patoolib  # to extract zip file
import pandas as pd

def load_data():
    Path("datasets").mkdir(parents=True, exist_ok=True) # creates a directory named datasets, here we will extract our data
    # Checking if the .csv file is already present, if not then only we will run the further code
    if not Path("datasets/life_expectancy.csv").is_file():
        # Using pattol library to exract our winRAR zip file
        patoolib.extract_archive("./archive.zip", outdir="datasets")
    # reading the file using pandas and returning it
    return pd.read_csv(Path("datasets/life_expectancy.csv"))
````

The dataset has been downloaded and extracted.

````python
life_exp = load_data()
````

### Now let's answer some question

* Answer the following questions among these countries only: among `["India", "Sri Lanka", "Pakistan", "Australia", "Canada", "Japan", "Bangladesh"]` <br><br>
Q1. Which country has highest life expectancy?  <br>
Q2. Which country has highest GDP per capita? <br>
Q3. Which country consumes Alcohol most and least? <br>
Q4. Which country has highest Literacy Rate? <br>
Q5. Which country has highest infection rate?


### To answer our questions let's compute the following first:

Let's take mean of `life expectancy`, `GDP`, `Schooling` and `Alcohol` from 2000 to 2015 for all countries, and plot:
* Country vs life expectancy
* Country vs GDP
* Country vs Alcohol consumption
* Country vs Schooling
* Coutry vs Infection rate 

````python
country_vs_lx = life_exp.groupby("Country")[["Life expectancy", "GDP", "Alcohol", "Schooling"]].mean()
country_vs_lx
````

#### Q1: Which country has highest life expectancy?

*Extracting selected rows only*

````python
country = ["India", "Sri Lanka", "Pakistan", "Australia", "Canada", "Japan", "Bangladesh"]
sel_country = country_vs_lx[country_vs_lx.index.isin(country)]
sel_country
````

### Selected `countries` vs `Life expectancy` Bar graph

````python
plt.title("Country vs Life expectancy")
plt.ylabel("Life Expectancy")
plt.xlabel("Country")
plt.bar(sel_country.index, sel_country["Life expectancy"]);
````

Here, we can conclude that `Japan` has the highest Avg Life expectancy followed by `Australia` from the selected countries.

#### Q2. Which country has highest GDP per capita?

Now, let's see selected `country` vs avg `GDP` pie chart

````python
plt.figure(figsize=(12, 8))
plt.title("Country vs GDP")
plt.pie(sel_country.GDP, labels=sel_country.index, autopct='%.0f%%');
````

Here, we can see that `Australia` has best GDP followed by `Canada`

#### Q3. Which country consumes Alcohol most and least?

````python
plt.title("Country vs Alcohol Consumption")
plt.xlabel("Country")
plt.ylabel("Alcohol Consumption")
plt.plot(sel_country.Alcohol, "b-x");
````

Here, we can conclude that, `Australia` has maximum alcohol consumption rate followed by `Canada`, <br>
While `Pakistan` and `Bangladesh` cosumes least Alcohol.

#### Q4 Which country has highest and lowest literacy rate?

*Here we are using `barplot` from `seaborn` library*

````python
plt.yticks(rotation=10)
sns.barplot(y=sel_country.index, x=sel_country.Schooling, color="gray", width=0.6);
````

From the graph we can conclude that `Australia` has the highest literacy rate and `Pakistan` has lowest literacy rate

#### Q5. Which country has highest infection rate?

*Infected rate column will contain sum of mean of all the infections such as `Hepatitis B`, `Measles`, `Polio`, `HIV/AIDS`, `Diphtheria`*

````python
country_inf = life_exp.groupby("Country")[["Hepatitis B", "Measles", "Polio", "HIV/AIDS", "Diphtheria"]].mean()
country_inf
````

````python
# Extract some selected countries only
sel_country = country_inf[country_inf.index.isin(country)]
sel_country
````

Here, we can see `Japan` has no infections of `Hepatitis B`, so we should change it's NaN value to 0, to continue with our calculations

````python
sel_country['Hepatitis B'] = sel_country['Hepatitis B'].fillna(0)
sel_country
````

````python
sel_country["Infected rate"] = sel_country["Hepatitis B"] + sel_country["Polio"] +sel_country["Measles"] + sel_country["HIV/AIDS"] + sel_country["Diphtheria"]
````

````python
sel_country
````

Let's plot `Country` vs `Infected Rate`

````python
plt.title("Country vs Infection rate")
plt.xlabel("Country")
plt.ylabel("Infection")
plt.plot(sel_country["Infected rate"], "r--o");
````

Here, we can see `India` has the highest Infection rate
